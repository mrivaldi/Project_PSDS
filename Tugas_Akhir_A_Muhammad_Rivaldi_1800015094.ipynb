#**Google Play Store Apps**

berikut adalah data yang saya ambil dari kaggle, dimana data ini adalah data Google Play Store Aps. saya berniat mencari popularitar Apps berdasarkan category Apps pada dataset

## Pertanyaan Analisis

1. Category manakah yang paling sering diinstall?
2. Category manakah yang paling jarang diinstall?

## Data Preparation

untuk data prepraretion, saya menyiapkan data frame untuk jenis-jenis Category beserta jumlah Install-nya

```
import pandas as pd
import plotly.express as px
```

```
#import file csv dari github
df = pd.read_csv('https://raw.githubusercontent.com/mrivaldi/file/main/googleplaystore.csv')
df
```

```
#melihat nama-nama data serta jumlahnya yang ada pada colom Type
df.Installs.value_counts()
```

```
#menghilangkan tanda '+' dan ',' pada kolom Installs
#mengubah data 'Free' menjadi '0' pada kolom Installs
#mengkonversi type data pada kolom Installs  menjadi int
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('Free', '0')
df['Installs'] = df['Installs'].astype(int)
df
```

```
#membuat data List yang berisi semua Category yang ada pada Dataset
List_category = list(pd.unique(df['Category']))
List_category
```

```
#membuat data List yang berisi Jumlah Installs Apps berdasarkan tiap Category
list_sum = []
for i in List_category:
  df_cat = df[df['Category'] == i]
  sum_ins = df_cat.Installs.sum()
  list_sum.append(sum_ins)

print(list_sum)
```

```
#membuat data frame dari jumlah install pada setiap category
df_install = pd.DataFrame({"Category":List_category,
              "Jumlah_Install":list_sum})
df_install.sort_values(['Jumlah_Install', 'Category'], ascending=[True, False], inplace=True)
df_install
```

## Data Visualization

```
#Visualisasi Data dengan Membuat Plot Bar berdasarkan Dataframe df_install
gbr_3 = px.bar(df_install,
               x = 'Category',
               y = 'Jumlah_Install')

gbr_3.update_layout(
    title = 'Diagram Batang jumlah instal Apps Pada Google Play Store',
    title_font_color = 'blue',
    title_font_size = 30,

    xaxis = dict(
        title = 'Category',
        titlefont_size = 16,
        tickfont_size = 9),
    
    yaxis = dict(
        title = 'Jumlah Install',
        titlefont_size = 16,
        tickfont_size = 14))

gbr_3.show()
```
